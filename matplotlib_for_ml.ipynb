```python
# importing atplotlib library
import matplotlib.pyplot as plt
# pyplot is nothing but python plot
```

```python
#import numpy and pandas
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0,10,100)
# it is present in the numpy library
# it returns equispaced values between 0 and 100
```

```python
print(x)
```

```python
y = np.sin(x)
z = np.cos(x)
print(y)
print(z)
```

```python
#plotting the data
# if we plot x and y , we get a sine curve
# if we plot x and x , we get a cos curve
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
# adding title, x-axis,y-axis labels
plt.plot(x,y)
plt.title('sine curve')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.plot(x,z)
plt.title('cosine curve')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# parabola
x = np.linspace(-10,10,20)
# 20 equally spaced values between -10 and 10
y = x**2
plt.plot(x,y)

plt.show()
print(x)
print(y)
#
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
x = np.linspace(-5,5,50)
y = x**2
plt.plot(x,np.sin(x),'gx')
plt.plot(x,np.cos(x),'r--')

plt.show()
```

```python
# barplot
fig = plt.figure()
# this creates an empty plot
ax = fig.add_axes([0,0,1,1])
# this adds axes to the empty plot
languages = ['eng','french','spanish','latin','german']
people = [100,50,40,57,68]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('number of people')
plt.title('languages and people')
plt.show()
```

```python
# pie chrt
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax1.pie(people,labels=languages,autopct='%1.2f%%')
plt.show()
```

```python
# scatter plot

x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax3 =  fig2.add_axes([0,0,1,1])
ax3.scatter(x,y,color='g')
ax3.scatter(x,z,color='r')
plt.show()
```

```python
# 3D scatterplot
fig2 == plt.figure()
ax4 = plt.axes(projection='3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax4.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
```
