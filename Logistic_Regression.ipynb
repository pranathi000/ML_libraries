3. def update_weights():

  * Step 1: weights and bias are randomly selected.
  * Step 2: Then the Z = wX+b is calculated.
  * step 3: the calculated Z is used to find the sigmoid function which is y_cap =  1/(1+e^(-z))
  * step 4: And then the cost function is calculated.
* If the cost function is high, then the weights and bias needs to be updated in order to maintain better accuracy and less loss value.
* This updating of weights and biases includes gradient Descent algorithm
* To help you revise what I told, We- upto now have discussed a total of two forumals where Z and sigmoid function need to be calculated.
* But in order to update weights and bias we need some differential calculus where we have keep in mind the 2 forumals.
* w  = w - (L)*(dw)

  b =  b - (L)*(db)
* where L is the learning rate. It tells on how much amount the weights and bias need to change in order to achieve optimal values and a good accuracy score.
* dw and db are partially differentiated with respect to the cost function.
* The dw and db formulas goes like this

  dw = 1/m*( (y_cap - y).X)

  db = 1/m* ((y_cap - y))
